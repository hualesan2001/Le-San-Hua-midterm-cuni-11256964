```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import random
%matplotlib inline
```

```python
np.random.seed(11256964)
```

```python
my_list = ("electronics", "clothing", "books", "food")
discount =  (0,0.1,0.2)
df = pd.DataFrame({"customer_id": np.random.randint (1,41,200),
                  "category": np.random.choice (my_list,200),
                  "price": np.abs(np.random.normal(50, 15, 200)),
                  "quantity" : np.random.randint (1,6,200),
                  "discount" : np.random.choice (discount,200),
                  "day": np.random.randint (1,31,200)})
```

```python
print (df)
```

```python
#Task 1
df["gross_revenue"]= df["price"] * df["quantity"]
df["net_revenue"] = df["gross_revenue"] * (1 - df["discount"])
print(df["gross_revenue"])
print (df["net_revenue"])
```

```python
#Task 2
df["category_avg_net_revenue"] = df.groupby("category")["net_revenue"].transform("mean")

df.head()
```

```python
#Task 3
df[(df["net_revenue"] > df["category_avg_net_revenue"])]
df.sample (7)
```

```python
#Task 4
def order_type(row):
    if row["quantity"] >= 4 and row["discount"] > 0:
        return "bulk_discounted"
    elif row["quantity"] >= 4 and row["discount"] == 0:
        return "bulk_fullprice"
    elif row["quantity"] < 4 and row["discount"] > 0:
        return "standard_discounted"
    else:
        return "standard_fullprice"

df["order_type"] = df.apply(order_type, axis=1)

print (df["order_type"])
```

```python
#Task 5
summary_table = df.groupby("category").agg({
    "price": "mean",
    "quantity": "sum",
    "discount": "mean",
    "net_revenue": "sum",
    "customer_id": "count"
})
```
