```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
n1 = np.random.randint(100, 501)
n2 = np.random.randint(100, 501)
n3 = np.random.randint(100, 501)
n4 = np.random.randint(100, 501)
```

```python
x1 = np.random.normal(loc=0, scale=1, size=n1)      
x2 = np.random.uniform(low=0, high=1, size=n2)      
x3 = np.random.exponential(scale=1, size=n3)          
x4 = np.random.poisson(lam=4, size=n4)           
print(n1, n2, n3, n4)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].hist(x1, bins=20)
axes[0, 0].set_title("Normal")
axes[0, 1].hist(x2, bins=20)
axes[0, 1].set_title("Uniform")
axes[1, 0].hist(x3, bins=20)
axes[1, 0].set_title("Exponential")
axes[1, 1].hist(x4, bins=20)
axes[1, 1].set_title("Poisson")
plt.tight_layout()
plt.show()
```
